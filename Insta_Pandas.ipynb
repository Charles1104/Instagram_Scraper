```python
the_item = 'mx5'
data_path = "data/data_photos_mx5/"
```

```python
f = open(data_path+"metadata.json","r")
the_posts = json.load(f)
```

```python
list_Media_ID = the_posts.keys()
```

```python
def get_list_User_ID(the_posts):
    list_User_ID = []
    for i in list_Media_ID:
        list_User_ID.append(the_posts[i]['shortcode_media']['owner']['id'])
    return list_User_ID
```

```python
def get_list_username(the_posts):
    list_username = []
    for i in list_Media_ID:
        list_username.append(the_posts[i]['shortcode_media']['owner']['username'])
    return list_username
```

```python
def get_list_userID_profile_pic(the_posts):
    list_userID_profile_pic = []
    for i in list_Media_ID:
        list_userID_profile_pic.append(the_posts[i]['shortcode_media']['owner']['profile_pic_url'])
    return list_userID_profile_pic
```

```python
def get_list_location(the_posts):
    list_location = []
    for i in list_Media_ID:
        if the_posts[i]['shortcode_media']['location'] is None:
            list_location.append("NA")
        else:
            list_location.append(the_posts[i]['shortcode_media']['location']['name'])
    return list_location
```

```python
def get_list_tot_comments(the_posts):
    list_tot_comments = []
    for i in list_Media_ID:
        list_tot_comments.append(the_posts[i]['activity'][-1]['comments'])
    return list_tot_comments
```

```python
def get_list_tot_likes(the_posts):
    list_tot_likes = []
    for i in list_Media_ID:
        list_tot_likes.append(the_posts[i]['activity'][-1]['likes'])
    return list_tot_likes
```

```python
def get_list_tot_vidviews(the_posts):
    list_tot_vidviews = []
    for i in list_Media_ID:
        list_tot_vidviews.append(the_posts[i]['activity'][-1]['video_views'])
    return list_tot_vidviews
```

```python
def get_list_ts(the_posts):
    list_ts = []
    for i in list_Media_ID:
        list_ts.append(the_posts[i]['activity'][-1]['scrape_ts'])
    return list_ts
```

```python
def get_caption_text(the_posts):
    list_of_captions = []
    for i in list_Media_ID:
        try:
            list_of_captions.append(the_posts[i]["shortcode_media"]["edge_media_to_caption"]["edges"][0]["node"].get("text",""))
        except:
            list_of_captions.append('No caption')
    return list_of_captions
```

```python
def get_list_hashtags_caption(the_posts):
    list_of_hastags = []
    for i in list_Media_ID:
        list_of_hastags.append(the_posts[i]['caption_tags'])
    return list_of_hastags
```

```python
def get_list_of_comments(the_posts):
    list_of_comments = []
    for i in list_Media_ID:
        list_of_comments.append(the_posts[i]["comm_texts"])
    return list_of_comments   
```

```python
def get_list_of_comments_User_ID(the_posts):
    list_of_comments_User_ID = []
    for i in list_Media_ID:
        list_per_post = []
        try:
            for a in the_posts[i]['shortcode_media']['edge_media_to_comment']['edges']:
                list_per_post.append(a['node']['owner']['id'])
            list_of_comments_User_ID.append(list_per_post)
        except:
            list_of_comments_User_ID.append("No Comments")
    return list_of_comments_User_ID
```

```python
def get_list_of_coms_hashtags(the_posts):
    list_of_coms_hashtags = []
    for i in list_Media_ID:
        list_of_coms_hashtags.append(the_posts[i]["comm_tags"])
    return list_of_coms_hashtags   
```

```python
def get_body_media(the_posts):
    list_of_body_media = []
    for i in list_Media_ID:
        list_of_body_media.append("https://www.instagram.com/p/" + i)
    return list_of_body_media
```

```python
def get_list_of_Media_Body_URL(the_posts):
    list_of_Media_Body_URL = []
    for i in list_Media_ID:
        list_of_Media_Body_URL.append(the_posts[i]["object_url"])
    return list_of_Media_Body_URL
```

```python
def get_list_of_file_name(the_posts):
    list_of_file_name = []
    for i in list_Media_ID:
        list_of_file_name.append(the_posts[i]["the_fname"])
    return list_of_file_name   
```

```python
#function to have a visual on all the data that can be retrieved from the master page of the hashtag
def total_posts(hashtag):
    tt = requests.get("https://www.instagram.com/explore/tags/" + hashtag,timeout=5)
    tt_soup = BeautifulSoup(tt.text, "lxml")
    all_texts = [u.get_text() for u in tt_soup.find_all("script")]
    all_texts = [at for at in all_texts if "window._sharedData" in at[:50]][0]
    all_texts = all_texts.replace("window._sharedData","")
    all_texts = all_texts.replace("true","True")
    all_texts = all_texts.replace("false","False")
    all_texts = all_texts.replace("null","None")
    info = literal_eval(all_texts[3:-1])
    
    return info['entry_data']['TagPage'][0]['graphql']['hashtag']['edge_hashtag_to_media']['count']
```

```python
#Initialize dataframe with desired columns
df = pd.DataFrame(columns=['Hashtag','tot_posts','User_ID','Username','Profile_Pic_URL','Media_ID','Picture_Location',\
                           'tot_comments','tot_likes','video_views','Main_Caption', 'List_Of_Hashtags_Caption',\
                           'List_Of_Users_ID','List_Of_Comments','List_0f_Hashtags_Comments','User_Bio',\
                           'Body_Media_URL','Media_URL','FileName','scrape_ts'],\
                  index = range(1,1001))
```

```python
#Load data into the different columns
df['Hashtag'] = the_item
df['tot_posts'] = total_posts(the_item)
df['User_ID'] = get_list_User_ID(the_posts)
df['Username'] = get_list_username(the_posts)
df['Profile_Pic_URL'] = get_list_userID_profile_pic(the_posts)
df['Media_ID'] = the_posts.keys()
df['Picture_Location'] = get_list_location(the_posts)
df['tot_comments'] = get_list_tot_comments(the_posts)
df['tot_likes'] = get_list_tot_likes(the_posts)
df['video_views'] = get_list_tot_vidviews(the_posts)
df['Main_Caption'] = get_caption_text(the_posts)
df['List_Of_Hashtags_Caption'] = get_list_hashtags_caption(the_posts)
df['List_Of_Users_ID'] = get_list_of_comments_User_ID(the_posts)
df['List_Of_Comments'] = get_list_of_comments(the_posts)
df['List_0f_Hashtags_Comments'] = get_list_of_coms_hashtags(the_posts)
# df['User_Bio'] =
df['Body_Media_URL'] = get_body_media(the_posts)
df['Media_URL'] = get_list_of_Media_Body_URL(the_posts)
df['FileName'] = get_list_of_file_name(the_posts)
df['scrape_ts'] = get_list_ts(the_posts)
```
